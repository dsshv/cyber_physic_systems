## The Callendar-Van Dusen equation

```python
A = 3.9083 * 10 ** -3
B = -5.7775 * 10 ** -7
C = -4.183 * 10 ** 12
R_zero = 100
```

```python
def calculate(t, a, b, c, r):
    print(t)
    if t < 0:
        result = r * (1 + (a * t) + (b * t ** 2) + (c * t ** 3)* (t - 100))
        return result
    else:
        result = r * (1 + a * t + (b * t **2 ))
        return result

R_t = np.asarray([calculate(t, A, B, C, R_zero) for t in diff])
```

```python
plt.plot(R_t[0:100])
```
