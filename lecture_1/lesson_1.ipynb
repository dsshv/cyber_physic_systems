# Линейный конгруэнтный генератор

```python
def seed(inited_seed):
    global rand
    rand = inited_seed

def LCG():
    a = 77232917
    c = 82589933
    m = 2**10
    global rand
    psevdo = (a*rand + c) % m
    return psevdo / m

def get_randoms(x):
    seed(x)
    nums = []
    for i in range(100000):
        new_num = LCG()
        nums.append(new_num)
        seed(nums[i])
    return nums

print(get_randoms(37))
```

```python
import matplotlib.pyplot as plt
```

```python
psevdo_random_array = get_randoms(37)
```

# Вероятностные распределения

### Равномерное распределение

```python
from scipy.stats import uniform
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(10,10)})
```

```python
n = 50000
start = 10
width = 200
data_uniform = uniform.rvs(size=n, loc = start, scale = width)
```

```python
ax = sns.displot(
    data_uniform,
    bins=10,
    kde=True,
    color='red',
)
ax.set(xlabel='Uniform_distribution', ylabel='frequency')
```

### Нормальное распределние

```python
from scipy.stats import norm
n = 50000
start = 0
width = 1
data_normal = norm.rvs(size=n, loc = start, scale = width)
```

```python
ax = sns.displot(
    data_uniform,
    bins=100,
    kde=True,
    color='purple',
)
ax.set(xlabel='Normal_distribution', ylabel='frequency')
```

### Гамма распределение

```python
from scipy.stats import gamma
n = 50000
centres = 2
width = 1
data_gamma = gamma.rvs(a=centres, scale = width, size = n)
```

```python
ax = sns.displot(
    data_gamma,
    bins=100,
    kde=True,
    color='green',
)
ax.set(xlabel='Gamma_distribution', ylabel='frequency')
```

## Экпетонциальное распределение

```python
from scipy.stats import expon
n = 50000
start = 0
width = 1
data_expon = expon.rvs(scale = width, loc = start,  size = n)
```

```python
ax = sns.displot(
    data_expon,
    bins=50,
    kde=True,
    color='yellow',
)
ax.set(xlabel='Expon_distribution', ylabel='frequency')
```

## Распределение Пуассона

```python
from scipy.stats import poisson
n = 50000
m = 3
data_poisson = poisson.rvs(mu = m,  size = n)
```

```python
ax = sns.displot(
    data_poisson,
    bins=30,
    kde=True,
    color='blue',
)
ax.set(xlabel='Poisson_distribution', ylabel='frequency')
```

## Биномиальное

```python
from scipy.stats import binom
n_num = 10
p_num = 0.8
s = 50000
data_binom = binom.rvs(n = n_num,  p = p_num, size = s)
```

```python
ax = sns.displot(
    data_binom,
    bins=30,
    kde=True,
    color='pink',
)
ax.set(xlabel='Poisson_distribution', ylabel='frequency')
```
